```python
import pandas
import json
import csv
```

```python
# Open, inspect, and load into variable
with open('Resources/case1.json', 'r') as json_file:
    case1_json = json.load(json_file)
    
print(json.dumps(case1_json, indent=1))
```

```python
# Create variable for json to csv column
case_1 = case1_json['values'] 

# Open blank csv for writing json into
case_1_csv = open('case1.csv', 'w')

# Create csv writer object
csv_writer = csv.writer(case_1_csv)

# Counter variable for writing headers into csv
count = 0

for x in case_1:
    if count == 0:
        # Write headers of csv file
        header = x.keys()
        csv_writer.writerow(header)
        count += 1
        
    # Writing data of csv file
    csv_writer.writerow(x.values())
    
case_1_csv.close()
```
